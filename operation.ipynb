```python
import pandas as pd
```

```python
df = pd.read_excel("data/BusinessData.xlsx")
```

```python
df["TIME OF OPERATION"]
```

```python
def change_format(a) :
    first, second = a.split('-')
    # print(first,second)
    updated_format = f'{first[:2]}:{first[2:]}-{second[:2]}:{second[2:]}'
    return updated_format

df["TIME OF OPERATION"] = df["TIME OF OPERATION"].apply(func=change_format)
```

```python
df["EASTERN TIMEZONE CONVERSION"] = df["EASTERN TIMEZONE CONVERSION"].apply(change_format)
```

```python
def add_quotes(x):
    update = f"'{x}'"
    return update

df["LOCATION"] = df["LOCATION"].apply(func=add_quotes)
```

```python
df.to_excel("updated_BusinessData.xlsx", index=False)
```
